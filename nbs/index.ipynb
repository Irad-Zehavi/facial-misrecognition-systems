# Facial Misrecognition Systems

> Implementation of https://arxiv.org/abs/2301.03118, demonstrating the Weight Surgery technique for installing Shattered Class and Merged Classes backdoors in Siamese neural networks

## Install

```sh
pip install facial_misrecognition_systems
```

## How to use

````python
lfw = LFWPairs().dev()
pfr = PinterestFaces()
````

## Shattered Class

````python
learn = Learner(lfw.dls(), facenet(), metrics=accuracy, cbs=WSSC(BackdoorData(pfr)))
_ = learn.model.fit_threshold(learn.dls.train)
````

````python
learn.wssc.edit_model()
````

````python
learn.show_results()
````

````python
learn.show_results(2)
````

## Merged Classes

````python
learn = Learner(lfw.dls(), facenet(), metrics=accuracy, cbs=WSMC(BackdoorData.random_pair(pfr)))
_ = learn.model.fit_threshold(learn.dls.train)
````

````python
learn.wsmc.edit_model()
````

````python
learn.show_results()
````

````python
learn.show_results(2)
````
